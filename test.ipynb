```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lognormal_list = []
exponential_list = []
for i in range(1000):
    lognormal_list.append(np.random.lognormal(mean=np.log(2), sigma=1))
    exponential_list.append(np.random.exponential(scale=1))

plt.hist(lognormal_list, bins=100, density=True)
plt.show()

plt.hist(exponential_list, bins=100, density=True)
plt.show()
```
